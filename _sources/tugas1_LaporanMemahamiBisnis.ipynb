## BAB I PENDAHULUAN

A. LATAR BELAKANG

Pasar saham merupakan salah satu instrumen investasi yang memiliki peran penting dalam perekonomian global maupun nasional. Fluktuasi harga saham yang dipengaruhi oleh berbagai faktor ekonomi, politik, dan sosial membuat prediksi pergerakan harga saham menjadi tantangan yang kompleks. Dalam konteks ini, kemampuan untuk memprediksi harga saham secara akurat memiliki nilai yang sangat strategis bagi investor, manajer portofolio, dan pembuat kebijakan ekonomi.
PT. Indofood Sukses Makmur Tbk (INDF) adalah salah satu perusahaan terkemuka di Indonesia yang bergerak di sektor makanan dan minuman. Sebagai perusahaan dengan kapitalisasi pasar yang besar dan likuiditas saham yang tinggi, saham INDF menjadi salah satu pilihan utama bagi investor di Bursa Efek Indonesia. Mengingat pentingnya saham INDF dalam portofolio investasi, pemahaman yang mendalam tentang dinamika pergerakan harga saham ini sangat diperlukan.
Analisis time series adalah salah satu pendekatan yang populer dalam memprediksi harga saham. Metode ini memungkinkan analisis terhadap pola data historis untuk memprediksi pergerakan harga di masa depan. Model Autoregressive Integrated Moving Average (ARIMA) merupakan salah satu metode time series yang sering digunakan karena kemampuannya untuk menangkap pola data stasioner dan mengakomodasi tren serta komponen musiman. Dengan menggunakan model ini, penelitian ini bertujuan untuk mengidentifikasi pola pergerakan harga saham INDF dan memprediksi harga di masa depan, yang diharapkan dapat memberikan wawasan bagi pengambilan keputusan investasi yang lebih baik.
Penelitian ini didorong oleh kebutuhan untuk menyediakan alat prediksi yang akurat dan andal bagi investor, serta berkontribusi pada pengembangan strategi perdagangan yang lebih efektif. Melalui prediksi yang dihasilkan, investor dapat meminimalkan risiko dan memaksimalkan keuntungan dalam berinvestasi di pasar saham.

B.	Rumusan Masalah

1.	Bagaimana karakteristik dan pola pergerakan harga saham PT. Indofood Sukses Makmur Tbk (INDF) berdasarkan data historis?
2.	Apakah model time series, khususnya model Autoregressive Integrated Moving Average (ARIMA), dapat digunakan untuk memprediksi harga saham INDF di masa depan dengan tingkat akurasi yang memadai?
3.	Faktor-faktor apa saja yang mempengaruhi tingkat akurasi prediksi harga saham INDF menggunakan metode analisis time series?
4.	Bagaimana hasil prediksi harga saham INDF dapat digunakan sebagai dasar untuk pengambilan keputusan investasi yang lebih baik oleh investor?

C.	Tujuan Proyek


1.	Menganalisis karakteristik dan pola pergerakan harga saham PT. Indofood Sukses Makmur Tbk (INDF) berdasarkan data historis.
2.	Mengidentifikasi model time series yang paling sesuai, terutama model ARIMA, untuk memprediksi harga saham INDF di masa depan.
3.	Mengevaluasi tingkat akurasi prediksi harga saham INDF menggunakan metode analisis time series.
4.	Memberikan rekomendasi kepada investor berdasarkan hasil prediksi harga saham INDF untuk membantu pengambilan keputusan investasi yang lebih efektif.

D.	Manfaat Proyek

1.	Manfaat Akademis
•	Penelitian ini dapat memperkaya literatur dalam bidang keuangan, khususnya mengenai penggunaan analisis time series dalam memprediksi harga saham di pasar modal.
•	Hasil penelitian ini diharapkan dapat menjadi referensi bagi akademisi dan peneliti lain yang ingin mengkaji lebih lanjut tentang prediksi harga saham dan model ARIMA.
2.	Manfaat Praktis
•	Bagi investor dan manajer portofolio, hasil penelitian ini dapat memberikan panduan dalam mengambil keputusan investasi berdasarkan prediksi pergerakan harga saham PT. Indofood Sukses Makmur.
•	Model prediksi yang dihasilkan dapat digunakan sebagai alat bantu dalam strategi investasi, membantu investor dalam mengurangi risiko dan memaksimalkan potensi keuntungan.   
3.	Manfaat bagi Perusahaan
•	Penelitian ini dapat memberikan wawasan kepada manajemen PT. Indofood Sukses Makmur mengenai persepsi pasar terhadap saham mereka, yang dapat dijadikan dasar untuk pengambilan keputusan strategis perusahaan.
4.	Manfaat bagi Regulator
•	Hasil penelitian ini dapat memberikan masukan kepada regulator pasar modal, seperti Otoritas Jasa Keuangan (OJK), terkait perkembangan metode prediksi yang dapat membantu menciptakan pasar yang lebih efisien dan stabil.
